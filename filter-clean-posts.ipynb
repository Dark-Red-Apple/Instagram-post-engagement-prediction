```python
# -------------------------------------------------------- New features
```

```python
# Extract emotion

from textblob import TextBlob

def detect_sentiment(txt):
    blob_text = TextBlob(txt)
    return blob_text.sentiment 


df_sorted['sentiment'] = df_sorted['caption_text_trimmed'].apply(lambda x: detect_sentiment(x))
# print(df_caption_text_trimmed['sentiment'])
```

```python
# Get the lenght of caption

df_sorted['caption_length'] = df_sorted['caption_text_trimmed'].apply(lambda x: len(x))
# print(df_caption_text_trimmed.sort_values('caption_length', ascending = False) )
```

```python
# Check if emojis are used

df_sorted['emojis_used'] = df_sorted['emojis'].apply(lambda x: not pd.isnull(x))
# print(df_caption_text_trimmed['emojis_used'])
```

```python
# checking if someone is tagged in caption

import re

pattern = '(?:^|[^\w])(?:@)([A-Za-z0-9_](?:(?:[A-Za-z0-9_]|(?:\.(?!\.))){0,28}(?:[A-Za-z0-9_]))?)'
df_sorted['tagged_accounts'] = df_sorted['caption_text_trimmed'].apply(lambda x: re.findall(pattern, x)) 
df_sorted['tagged_accounts'] = df_sorted['tagged_accounts'].apply(lambda x: len(x)) 
```

```python
print(df_sorted[['tagged_accounts', 'caption_length', 'emojis_used', 'sentiment']].head())
```

```python
# -------------------------------------------------------- normalize
```

```python
# seperate emotions  
df_sorted.reindex()
df_sorted[['polarity', 'subjectivity']] = pd.DataFrame(df_sorted['sentiment'].tolist(), index=df_sorted.index)                                                                                                                       
                                                                                                
```

```python
from sklearn.preprocessing import OneHotEncoder

#creating instance of one-hot-encoder
encoder = OneHotEncoder(handle_unknown='ignore')

#perform one-hot encoding on 'team' column 
encoder_df = pd.DataFrame(encoder.fit_transform(df_sorted[['__typename']]).toarray(), index=df_sorted.index)

print(encoder_df)
#merge one-hot encoded columns back with original DataFrame
df_sorted = df_sorted.join(encoder_df)
```
